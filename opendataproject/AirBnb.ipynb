```python
airbnb_data["id"].plot(kind="bar")
plt.title("Case IDs", fontsize=26)
plt.xlabel("Case", fontsize=26)
plt.ylabel("ID", fontsize=26)
```

```python
airbnb_data["host_id"].plot(kind="bar")
plt.title("AirBnb Host Id Numbers", fontsize=26)
plt.xlabel("Host", fontsize=26)
plt.ylabel("ID", fontsize=26)
```

```python
airbnb_data["host_name"].value_counts().plot(kind="bar")
plt.title("Host Names", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
name_counts = airbnb_data["host_name"].value_counts()
name_counts[name_counts > 10].plot(kind="bar")
plt.title("Host Names", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
name_counts[name_counts > 50].plot(kind="bar", grid=True)
plt.title("Host Names", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
airbnb_data["neighbourhood"].value_counts().plot(kind="barh", grid=True)
plt.title("Neighbourhoods the Properties are Located", fontsize=26)
plt.xlabel("Properties", fontsize=26)
plt.ylabel("Cities/Suburbs", fontsize=26)
```

```python
airbnb_data["latitude"].hist()
plt.title("Latitude of Accommodations", fontsize=26)
plt.xlabel("Latitudinal Value", fontsize=26)
plt.ylabel("Property Number", fontsize=26)
```

```python
airbnb_data["longitude"].hist()
plt.title("Longitude of Accommodations", fontsize=26)
plt.xlabel("Longitudinal Value", fontsize=26)
plt.ylabel("Property Number", fontsize=26)
```

```python
airbnb_data["room_type"].value_counts().plot(kind="pie")
plt.title("Types of Rooms offered by Airbnb", fontsize=26)
plt.ylabel("", fontsize=26)
```

```python
price_bar = airbnb_data["price"].value_counts()
price_bar[price_bar < 4000].plot(kind="bar")
plt.title("Prices per Night", fontsize=26)
plt.xlabel("Prices in $", fontsize=26)
plt.ylabel("No. of Properties Available", fontsize=26)
```

```python
price_bar = airbnb_data["price"].value_counts()
price_bar[price_bar < 100].plot(kind="bar")
plt.title("Prices per Night", fontsize=26)
plt.xlabel("Prices in $", fontsize=26)
plt.ylabel("No. of Properties Available", fontsize=26)
```

```python
price_bar = airbnb_data["price"].value_counts()
price_bar[price_bar > 100].plot(kind="bar", grid=True)
plt.title("Prices per Night", fontsize=26)
plt.xlabel("Prices in $", fontsize=26)
plt.ylabel("No. of Properties Available", fontsize=26)
```

```python
price_bar = airbnb_data["price"].value_counts()
price_bar[price_bar > 500].plot(kind="bar", grid=True)
plt.title("Prices per Night", fontsize=26)
plt.xlabel("Prices in $", fontsize=26)
plt.ylabel("No. of Properties Available", fontsize=26)
```
