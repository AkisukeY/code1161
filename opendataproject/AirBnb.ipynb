```python
airbnb_data["id"].plot(kind="bar")
plt.title("Case IDs", fontsize=26)
plt.xlabel("Individual", fontsize=26)
plt.ylabel("ID", fontsize=26)
```

```python
zoom_id = airbnb_data["id"].value_counts()
zoom_id[zoom_id].plot(kind="bar")
plt.title("Case IDs", fontsize=26)
plt.xlabel("Individual", fontsize=26)
plt.ylabel("ID", fontsize=26)
```

```python
airbnb_data["name"].value_counts().plot(kind="bar")
```

```python
airbnb_data["host_id"].hist()
plt.title("AirBnb Host Id Numbers", fontsize=26)
plt.xlabel("ID", fontsize=26)
plt.ylabel("Host", fontsize=26)
```

```python
airbnb_data["host_name"].value_counts().plot(kind="bar")
plt.title("Host Names", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
name_counts = airbnb_data["host_name"].value_counts()
name_counts[name_counts > 10].plot(kind="bar")
plt.title("Host Names", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
name_counts[name_counts > 50].plot(kind="bar", grid=True)
plt.title("Host Names", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
airbnb_data["neighbourhood"].value_counts().plot(kind="bar")
plt.title("Neighbourhoods the Properties are Located", fontsize=26)
plt.xlabel("Neighbourhood", fontsize=26)
plt.ylabel("Property", fontsize=26)
```

```python
airbnb_data["latitude"].hist()
plt.title("Latitude of Accommodations", fontsize=26)
plt.xlabel("Latitude", fontsize=26)
plt.ylabel("Location", fontsize=26)
```

```python
airbnb_data["longitude"].hist()
plt.title("Longitude of Accommodations", fontsize=26)
plt.xlabel("Longitude", fontsize=26)
plt.ylabel("Location", fontsize=26)
```

```python
airbnb_data["room_type"].value_counts().plot(kind="bar")
plt.title("Types of Rooms offered by Airbnb", fontsize=26)
plt.xlabel("Type", fontsize=26)
plt.ylabel("Rooms", fontsize=26)
```

```python
airbnb_data["price"].hist()
plt.title("Prices", fontsize=26)
plt.xlabel("Amount in $", fontsize=26)
plt.ylabel("Location", fontsize=26)
```

```python
airbnb_data["minimum_nights"].hist()
plt.title("Minimum Nights", fontsize=26)
plt.xlabel("Number of Nights", fontsize=26)
plt.ylabel("Bookings", fontsize=26)
```

```python
airbnb_data["number_of_reviews"].hist()
plt.title("Number of ", fontsize=26)
plt.xlabel("Name", fontsize=26)
plt.ylabel("Property Owned", fontsize=26)
```

```python
review = airbnb_data["last_review"].value_counts()
review[review > 50].plot(kind="bar")
plt.title("Reviews at the End of the Month", fontsize=26)
plt.xlabel("Date", fontsize=26)
plt.ylabel("Reviews", fontsize=26)
```

```python
airbnb_data["reviews_per_month"].hist()
plt.title("Reviews Per Month", fontsize=26)
plt.xlabel("Number of Reviews", fontsize=26)
plt.ylabel("Reviewers", fontsize=26)
```

```python
airbnb_data["calculated_host_listings_count"].hist()
plt.title("Host Listings", fontsize=26)
plt.xlabel("Count", fontsize=26)
plt.ylabel("Listings", fontsize=26)
```

```python
airbnb_data["availability_365"].hist()
plt.title("Availability of Properties", fontsize=26)
plt.xlabel("Number of Days", fontsize=26)
plt.ylabel("Properties", fontsize=26)
```
